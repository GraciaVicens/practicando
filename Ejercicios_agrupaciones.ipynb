1) Usando el fichero WHO.csv, ¿Cuál es el volumen total de CO2 emitido por cada continente?

```python
import pandas as pd
import numpy as np

data = pd.read_csv("Data/WHO.csv")
data
byco2 = data.groupby("Continent").aggregate({
    "Total_CO2_emissions":np.sum})
byco2
```

4) Calcula la cantidad de ayuda recibida por cada municipio en función del númeto total de habitantes. (v2)

```python
import pandas as pd
import random

random.seed(0)

nombres = [f'Municipio{i}' for i in range(1, 11)]

data_municipios = {
    'Nombre': nombres,
    'Código Postal': [random.randint(10000, 99999) for _ in range(10)],
    'Población': [random.randint(1000, 50000) for _ in range(10)]  # Añadimos un atributo aleatorio, en este caso "Población"
}

df_municipios = pd.DataFrame(data_municipios)


data_ayudas = {
    'Nombre': [random.choice(nombres) for _ in range(20)],
    'Ayuda Económica (en euros)': [random.randint(1000, 10000) for _ in range(20)],
    'Número de Beneficiarios': [random.randint(10, 100) for _ in range(20)]
}

df_ayudas = pd.DataFrame(data_ayudas)

print("Dataframe de Municipios:")
print(df_municipios)

print("\nDataframe de Ayudas:")
print(df_ayudas)
```

```python
df_ayudas2 = df_ayudas.groupby(["Nombre"]).aggregate({
    "Ayuda Económica (en euros)":np.sum})
df_ayudas2
```

```python
new_data = df_ayudas2.merge(df_municipios, left_on="Nombre", right_on="Nombre")
new_data['relacion_ayuda_habitantes'] = new_data["Ayuda Económica (en euros)"] / new_data["Población"]
ordenado = new_data.sort_values(by="relacion_ayuda_habitantes", ascending=False)
ordenado
```

5) Agrupa los datos por el nombre de de la escuela. - ¿Qué escuela tiene más infantes? - ¿Qué escuela tiene los infantes más altos?

```python
data = {
    'school': ['s001', 's002', 's003', 's001', 's002', 's004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_Of_Birth': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'age': [12, 12, 13, 13, 14, 12],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']
}

df = pd.DataFrame(data, index=['S1', 'S2', 'S3', 'S4','S5','S6'])
df
```

```python
grupo1 = df.groupby(["school"])["name"].count().reset_index(name='Conteo')
valor_maximo = grupo1["Conteo"].max()
filas_valor_maximo = grupo1[grupo1['Conteo'] == valor_maximo]["school"].values[:]
print("Las escuelas con mas infantes son: ",filas_valor_maximo)
```

```python
grupo2 = df.groupby(["school"]).aggregate({
    "height":np.mean})
valor_mas_altos = grupo2["height"].max()
filas_mas_altos = grupo2[grupo2["height"] == valor_mas_altos]
nombre = filas_mas_altos.index[0]
print("la escuela con los infantes mas altos es: ", nombre)
```

6) Dado el siguiente listado de ventas: - ¿Qué comprador ha gastado más? - ¿Qué vendedor ha hecho más ventas?

```python
import pandas as pd
import numpy as np

# Crear datos aleatorios
np.random.seed(1)
n = 15
productos = ['Producto ' + str(i) for i in range(1, n+1)]
precios = np.random.randint(10, 100, n)
compradores = np.random.choice(['Juan', 'Pedro', 'Maria', 'Ana'], n)
vendedores = np.random.choice(['Carlos', 'Laura', 'Miguel', 'Elena'], n)

# Crear DataFrame
df_ventas = pd.DataFrame({
    'Producto': productos,
    'Precio': precios,
    'Comprador': compradores,
    'Vendedor': vendedores
})

print(df_ventas)
```

```python
bycomprador = df_ventas.groupby(["Comprador"]).aggregate({
    "Precio":np.sum})
maximo=bycomprador["Precio"].max()
filas_maximo = bycomprador[bycomprador['Precio'] == maximo]
nombre = filas_maximo.index[0]
print(nombre)
```

```python
byvendedor = df_ventas.groupby(["Vendedor"])["Precio"].count().reset_index(name="Conteo")
valor_maximo = byvendedor["Conteo"].max()
filas_valor_maximo = byvendedor[byvendedor['Conteo'] == valor_maximo]["Vendedor"].values[:]
print(filas_valor_maximo)
```

7) Fichero “data/data_groups_cursos.csv”; contiene cursos realizados por personas. Cada muestra corresponde a un curso. Se pide un listado del número de cursos realizado por cada persona y una valoración de sus puntos según el tipo de curso realizado: - ‘A’: 3 puntos - ‘B’: 2 puntos - ‘C’: 1 punto - ‘D’: 0.5 punto - “E”: 0.5 punto - “F”: 1 punto

```python
cursos = pd.read_csv("Data/data_groups_cursos.csv")
cursos
```

```python
puntos_diccionario = {'A': 3, 'B': 2, 'C': 1, 'D': 0.5, 'E': 0.5, 'F': 1}
cursos["valoracion"] = cursos["Curso"].map(puntos_diccionario)
cursos
```

```python
resultados_grupo = cursos.groupby('Nom').agg({'Nom': 'count', 'valoracion': 'sum'}).rename(columns={"Nom":"numero_cursos", "valoracion": 'suma_valoracion'})
resultados_grupo
```
